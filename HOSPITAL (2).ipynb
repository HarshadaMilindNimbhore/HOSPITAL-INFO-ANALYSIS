```python
def count_emergency_services(df, hospital_type):
    # Filter the DataFrame for the specified hospital type
    filtered_df = df[df['Hospital Type'] == hospital_type]
    
    # Count the number of hospitals providing emergency services
    emergency_count = filtered_df[filtered_df['Emergency Services'] == 'Yes'].shape[0]
    
    # Get the counts of emergency services for all hospital types for the bar plot
    emergency_counts = df[df['Emergency Services'] == 'Yes']['Hospital Type'].value_counts()

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(emergency_counts.index, emergency_counts.values, color='skyblue')
    plt.xlabel('Hospital Type')
    plt.ylabel('Count of Hospitals with Emergency Services')
    plt.title('Count of Hospitals Providing Emergency Services by Type')
    plt.xticks(rotation=45)
    plt.show()
    
    return emergency_count

# Taking input from the user
hospital_type = input("Enter the hospital type (Childrens, Critical Access Hospitals, Acute Care Hospitals): ")

# Check if the input is valid
if hospital_type not in ['Childrens', 'Critical Access Hospitals', 'Acute Care Hospitals']:
    print("Invalid hospital type entered. Please enter one of the following: Childrens, Critical Access Hospitals, Acute Care Hospitals.")
else:
    emergency_count = count_emergency_services(df, hospital_type)
    print(f'There are {emergency_count} hospitals of type "{hospital_type}" providing emergency services.')
```

# Distribution On the basis Of Hospital Type And their rating

```python
def count_high_rating_hospitals(df):
    # Taking input from the user
    hospital_type = input("Enter the hospital type (Childrens, Critical Access Hospitals, Acute Care Hospitals): ")

    # Check if the input is valid
    if hospital_type not in ['Childrens', 'Critical Access Hospitals', 'Acute Care Hospitals']:
        print("Invalid hospital type entered. Please enter one of the following: Childrens, Critical Access Hospitals, Acute Care Hospitals.")
        return

    # Define high rating threshold
    high_rating_threshold = 4

    # Filter the DataFrame for the specified hospital type
    filtered_df = df[df['Hospital Type'] == hospital_type]
    
    # Count the number of hospitals with high ratings
    high_rating_count = filtered_df[filtered_df['Hospital overall rating'] >= high_rating_threshold].shape[0]
    
    # Get the counts of high rating hospitals for all hospital types for the bar plot
    high_rating_counts = df[df['Hospital overall rating'] >= high_rating_threshold]['Hospital Type'].value_counts()

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.bar(high_rating_counts.index, high_rating_counts.values, color='skyblue')
    plt.xlabel('Hospital Type')
    plt.ylabel('Count of High-Rating Hospitals')
    plt.title('Count of High-Rating Hospitals by Type')
    plt.xticks(rotation=45)
    plt.show()
    
    print(f'There are {high_rating_count} hospitals of type "{hospital_type}" with a high rating.')

# Run the function
count_high_rating_hospitals(df)
```

# Statewise Distribution Of Hospitals Type 

```python
def count_hospital_by_state(df):
    # Taking input from the user
    hospital_type = input("Enter the hospital type (Childrens, Critical Access Hospitals, Acute Care Hospitals): ")

    # Check if the input is valid
    if hospital_type not in ['Childrens', 'Critical Access Hospitals', 'Acute Care Hospitals']:
        print("Invalid hospital type entered. Please enter one of the following: Childrens, Critical Access Hospitals, Acute Care Hospitals.")
        return

    # Filter the DataFrame for the specified hospital type
    filtered_df = df[df['Hospital Type'] == hospital_type]
    
    # Count hospitals by state
    state_counts = filtered_df['State'].value_counts()
    
    # Find the state with the highest count
    max_state = state_counts.idxmax()
    max_count = state_counts.max()

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.bar(state_counts.index, state_counts.values, color='skyblue')
    plt.xlabel('State')
    plt.ylabel('Count of Hospitals')
    plt.title(f'State-wise Count of Hospitals for {hospital_type}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    print(f'For "{hospital_type}", the state with the highest count of hospitals is {max_state} with {max_count} hospitals.')

# Run the function
count_hospital_by_state(df)
```
